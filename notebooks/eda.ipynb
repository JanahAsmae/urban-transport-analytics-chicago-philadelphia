<T1>Chargement & Structure Initiale des donnees</T1>

```python
# chargement des fichiers CSV

# r√©pertoire des donn√©es brutes :
data_dir = Path("..") / "data" / "raw"
csv_files = sorted([p for p in data_dir.iterdir() if p.suffix.lower() in [".csv", ".xlsx"]])

# liste de DataFrame (m√™me ordre que csv_files) et dict {stem: df} :
dfs = [pd.read_csv(p) if p.suffix.lower() == ".csv" else pd.read_excel(p) for p in csv_files]
dfs_names = [p.stem for p in csv_files]
dfs_dict = dict(zip(dfs_names, dfs))
```

```python
# Afficher un aper√ßu de chaque DataFrame
for name, df in dfs_dict.items():
    print(f"--- {name} ---")
    display(df.head())
```

```python
# extraire jour/mois/ann√©e de service_date
for name, df in dfs_dict.items():
    if 'service_date' in df.columns:
        df['service_date'] = pd.to_datetime(df['service_date'], errors='coerce')
        df['Calendar_Year'] = df['service_date'].dt.year
        df['Calendar_Month'] = df['service_date'].dt.month
        df['Calendar_Day'] = df['service_date'].dt.day
```

```python
# Taille des DataFrames
for name, df in dfs_dict.items():
    print(f" - {name}: {df.shape}")
```

```python
# noms des colonnes des DataFrames
for name, df in dfs_dict.items():
    print(f"Colonnes de {name}:\n {df.columns.tolist()}\n")
```

```python
# type des colonnes des DataFrames
for name, df in dfs_dict.items():
    print(f"Types des colonnes de {name}:\n{'*'*50}\n{df.dtypes}\n")
```

#### üö© **Remarque:**
- colonne date dans Avg_By_Route_Chicago est de type object il faut la convertir en datetime

```python
#------------------- Conversion de la colonne Date en datetime
df = dfs_dict.get('Avg_By_Route_Chicago')

df['Date'] = pd.to_datetime(
    df['Date'],
    format='%Y-%m-%dT%H:%M:%S',
    errors='raise'   # garantit qu‚Äôaucune valeur invalide ne passe silencieusement
)

# Suppression de la partie temps (on garde uniquement la date)
df['Date'] = df['Date'].dt.date

# (Optionnel mais propre) Reconvertir en datetime64[ns] sans heure
df['Date'] = pd.to_datetime(df['Date'])


#------------------- Extraire jour/mois/ann√©e de Date
for name, df in dfs_dict.items():
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df['Calendar_Year'] = df['Date'].dt.year
        df['Calendar_Month'] = df['Date'].dt.month
        df['Calendar_Day'] = df['Date'].dt.day
```

```python
# periode des donn√©es dans chaque DataFrame
for name, df in dfs_dict.items():
    if 'Calendar_Year' in df.columns:
        print(f"P√©riode des donn√©es dans {name}: {df['Calendar_Year'].min()} - {df['Calendar_Year'].max()}")
```

```python
## üö© Remarque :
# La p√©riode des donn√©es analys√©es est 2019 ‚Äì 2025.
# La p√©riode des donn√©es dans 'Avg_By_Mode_Chicago' et 'Avg_By_Route_Chicago' est 2001 ‚Äì 2025.

## üìå D√©cision
# Supprimer les donn√©es ant√©rieures √† 2019.
dfs_dict['Avg_By_Mode_Chicago'] = dfs_dict['Avg_By_Mode_Chicago'][dfs_dict['Avg_By_Mode_Chicago']['Calendar_Year'] >= 2019]
dfs_dict['Avg_By_Route_Chicago'] = dfs_dict['Avg_By_Route_Chicago'][dfs_dict['Avg_By_Route_Chicago']['Calendar_Year'] >= 2019]

# enregistre les modifications
dfs_dict['Avg_By_Mode_Chicago'].to_csv(data_dir / "Avg_By_Mode_Chicago_2019_2025.csv", index=False)
dfs_dict['Avg_By_Route_Chicago'].to_csv(data_dir / "Avg_By_Route_Chicago_2019_2025.csv", index=False)

# modification du dict pour refl√©ter le changement
dfs_dict['Avg_By_Mode_Chicago'] = pd.read_csv(data_dir / "Avg_By_Mode_Chicago_2019_2025.csv")
dfs_dict['Avg_By_Route_Chicago'] = pd.read_csv(data_dir / "Avg_By_Route_Chicago_2019_2025.csv")

display(dfs_dict['Avg_By_Mode_Chicago'].head())
print(f" - Avg_By_Mode_Chicago: {dfs_dict['Avg_By_Mode_Chicago'].shape}")
display(dfs_dict['Avg_By_Route_Chicago'].head())
print(f" - Avg_By_Route_Chicago: {dfs_dict['Avg_By_Route_Chicago'].shape}")
```

<T1> Analyse de Qualit√© des Donn√©es</T1>

```python
# valeurs manquantes par colonne de chaque DataFrame
for name, df in dfs_dict.items():
    valeurs_nulles = df.isnull().sum()
    taux_valeurs_nulles = round((df.isnull().sum() / len(df)) * 100, 2)

    df_nulles = pd.DataFrame({
        'Colonne': valeurs_nulles.index,
        'Nb_Manquants': valeurs_nulles.values,
        'Pourcentage': taux_valeurs_nulles.values
    }).sort_values('Nb_Manquants', ascending=False)
    
    print(f"--- Valeurs manquantes dans {name} ---")
    print('='*50,'\n',df_nulles,'\n')
```

```python
# valeurs doublons dans chaque DataFrame
for name, df in dfs_dict.items():
    nb_doublons = df.duplicated().sum()
    print(f"Nombre de lignes dupliqu√©es dans {name}: {nb_doublons}")
```

```python
# valeurs n√©gatives ou nulles dans la colonne 'Average_Daily_Ridership' de chaque DataFrame
for name, df in dfs_dict.items():
    if 'Average_Daily_Ridership' in df.columns:
        print(f"--{name}---")
        display(df[df['Average_Daily_Ridership'] <= 0])
    else :
        print(f"--{name}---")
        display(df[df['total_rides'] <= 0])
```
